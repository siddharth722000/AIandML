```python
#Question 2: Plot a pie chart for the country_name column
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
val=df.groupby('Country_Name').size()
v=list(val)
c=list(val.index)
plt.pie(v,labels=c,autopct="%0.1f%%",radius=1.5,
       colors=["red","blue","green",],
explode=[0.1,0,0],
       startangle=45)
plt.show()
```
