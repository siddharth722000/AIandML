```python
new_df=df.replace({'temperature':-99999, 'windspeed':-99999, 'event':'0'},np.nan)
new_df
```

## Imputation: It is a process of substituting the missing values

```python
#fullna() is a function which is used to fill the missing values
data=new_df.fillna(0) # press shift_tab 4 times for doccumentation of fillna()
data
```

```python
new_df
```

```python
f_data = new_df.fillna(method='ffill') #in this ffill method the nan values gets replaced by previous values
f_data
```

```python
# bfill--backfill
b_data=new_df.fillna(method='bfill')
b_data
```

```python
new_df
```

```python
new_df.mean()
```

```python
mean_data = new_df.fillna(new_df.mean()) #fill the missing values in a column with the columns mean value
mean_data
```

```python
new_df.median()
```

```python
median_data=new_df.fillna(new_df.median())
median_data
```

```python
new_df.event.mode().iloc[1]
```

```python
new_df['event']=new_df['event'].fillna(new_df.event.mode().iloc[0])
new_df
```
